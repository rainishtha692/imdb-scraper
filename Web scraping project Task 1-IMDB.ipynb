```python
from selenium import webdriver  # Import Selenium WebDriver
from selenium.webdriver.common.by import By  # Import By for locating elements
from selenium.webdriver.chrome.options import Options  # Import Chrome Options
from selenium.webdriver.chrome.service import Service  # Import Service to handle ChromeDriver
import pandas as pd  # Import pandas for data handling
import time  # Import time for sleep delays

# Path to your ChromeDriver executable
chrome_driver_path = r"C:\Users\Acer\Documents\chromedriver-win64\chromedriver.exe"

# Set up Chrome WebDriver options
options = Options()
options.add_argument('--headless')  # Run Chrome in headless mode (no UI)
options.add_argument('--disable-blink-features=AutomationControlled')  # Bypass bot detection
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36")  # Set a real user-agent to avoid blocking

# Initialize the Chrome WebDriver with the specified options
driver = webdriver.Chrome(service=Service(chrome_driver_path), options=options)

# IMDb Top 250 Movies page URL
url = 'https://www.imdb.com/chart/top/'
driver.get(url)  # Open the IMDb page in the browser

# Wait for the page to fully load
time.sleep(10)  # Adjust time if needed (ensures JavaScript content is loaded)

# Locate all movie entries (list items containing movie details)
movies = driver.find_elements(By.XPATH, '//li[contains(@class,"ipc-metadata-list-summary-item")]')

# Initialize an empty list to store movie data
movie_data = []

# Loop through the first 100 movies (modify if needed)
for movie in movies[:100]:
    try:
        # Extract the movie title without the ranking number (e.g., "1. The Shawshank Redemption" → "The Shawshank Redemption")
        title_raw = movie.find_element(By.XPATH, './/div[contains(@class,"ipc-title")]/a/h3').text
        title = title_raw.split('. ', 1)[-1]  # Splitting on ". " to remove the number

        # Extract the movie's release year
        year = movie.find_element(By.XPATH, './/div[contains(@class,"cli-title-metadata")]/span[contains(@class,"cli-title-metadata-item")][1]').text

        # Extract the IMDb rating (only the first value to exclude vote count)
        rating = movie.find_element(By.XPATH, './/span[contains(@class,"ipc-rating-star--rating")]').text.split()[0]

        # Append the extracted data to the list
        movie_data.append({
            'Title': title,
            'Year': year,
            'Rating': rating
        })
    
    except Exception as e:  # Catch and print errors without stopping execution
        print(f"Skipping a movie due to error: {e}")

# Quit the WebDriver after scraping is complete
driver.quit()

# Convert the movie data into a Pandas DataFrame
df = pd.DataFrame(movie_data)

# Print the first 20 rows to verify extracted data
print(df.head(20))

# Define the CSV file path to save the extracted data
csv_path = r'C:\Users\Acer\Documents\BDA3.3\Project 1- Web scraping\imdb_top_100_movies.csv'

# Save the DataFrame as a CSV file (without the index column)
df.to_csv(csv_path, index=False)

# Print a success message with the file location
print(f"✅ IMDb Top 100 Movies saved successfully at: {csv_path}")
```
